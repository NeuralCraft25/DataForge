```python
categories = ["Total", "european", "Maori", "Pacifica", "Asian", "Other", "Unknown"]

dataframes = {}
for category in categories:
    columns = ["Period"] + [col for col in df.columns if category in col]
    dataframes[category] = df[columns]

Maori_data = dataframes["Maori"]
european_data = dataframes[categories[1]]
Pacifica_data = dataframes["Pacifica"]
asian_data = dataframes["Asian"]
other_data = dataframes["Other"]
unknown_data = dataframes["Unknown"]
```

The **Corrections Amendment Act 2019** was passed to make several adjustments to the **Corrections Act 2004**, aiming to improve the management and rehabilitation of prisoners.
https://www.legislation.govt.nz/act/public/2019/0057/46.0/LMS23141.html

### **Why Was It Changed in 2019?**
- **Focus on rehabilitation and reintegration**: The government aimed to improve support for prisoners’ rehabilitation and reintegration into society, reducing the likelihood of reoffending.
- **Improved safety and wellbeing**: There were growing concerns about prison safety, prisoner welfare, and the effectiveness of rehabilitation programs.

### **What Was Changed?**
1. **Increased focus on rehabilitation**: New programs and services were introduced to help prisoners prepare for reintegration into society, including enhanced **mental health services** and **education**.
2. **Prison safety**: The Act made provisions to improve **security** and **safety** for both prisoners and staff.
3. **Enhanced monitoring**: More detailed monitoring of prison conditions and prisoner progress was implemented to ensure that rehabilitation and security were being maintained.
4. **Long-term offenders**: The law provided additional support for managing **long-term prisoners**, aiming to reduce the risk of violence and reoffending.

### **Why Did They Feel Another Change Was Needed?**
- The 2019 amendments were driven by ongoing concerns about the **effectiveness** of New Zealand’s corrections system, particularly in relation to **rehabilitation**, **recidivism**, and **prison overcrowding**.
- The government wanted to **improve outcomes** for prisoners, focusing not just on punishment, but on their **successful reintegration** into society.
